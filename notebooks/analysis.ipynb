```python
import sys, pandas as pd, numpy as np
import matplotlib.pyplot as plt, seaborn as sns
sys.path.insert(0, '..')
from filters import load_data
sns.set_theme(style='darkgrid'); plt.rcParams['figure.figsize']=(11,5)
df = load_data('../data/GlobalForestWatch_TreeCoverLoss.csv')
print('Shape:', df.shape)
df.head()
```

```python
print(df.dtypes)
print('\nMissing:\n', df.isnull().sum())
```

```python
df.describe().round(2)
```

```python
# Line chart — global trend
g = df.groupby('Year')['TreeCoverLoss_ha'].sum().reset_index()
plt.plot(g['Year'], g['TreeCoverLoss_ha']/1e3, color='#FF7043', lw=2.5, marker='o', ms=4)
plt.fill_between(g['Year'], g['TreeCoverLoss_ha']/1e3, alpha=0.2, color='#FF7043')
plt.title('Global Annual Tree Cover Loss 2001–2023'); plt.xlabel('Year'); plt.ylabel('Loss (K ha)')
plt.tight_layout(); plt.show()
```

```python
# Bar chart — top 10 countries
top = df.groupby('Country')['TreeCoverLoss_ha'].sum().nlargest(10).sort_values()
plt.barh(top.index, top.values/1e3, color='#4CAF50')
plt.title('Top 10 Countries by Total Tree Cover Loss'); plt.xlabel('Loss (K ha)')
plt.tight_layout(); plt.show()
```

```python
# Histogram
df['TreeCoverLoss_ha'].plot.hist(bins=40, color='#4CAF50', edgecolor='white', lw=0.3)
plt.title('Distribution of Annual Tree Cover Loss'); plt.xlabel('Loss (ha)')
plt.tight_layout(); plt.show()
```

```python
# Pie chart — drivers
drv = df.groupby('PrimaryDriver')['TreeCoverLoss_ha'].sum()
plt.pie(drv, labels=drv.index, autopct='%1.1f%%', startangle=130)
plt.title('Loss by Deforestation Driver'); plt.tight_layout(); plt.show()
```

```python
# Heatmap
cols = ['TreeCoverLoss_ha','TreeCoverGain_ha','FireArea_ha','CarbonLoss_tCO2','TreeCoverPct']
sns.heatmap(df[cols].corr(), annot=True, fmt='.2f', cmap='RdYlGn_r')
plt.title('Correlation Heatmap'); plt.tight_layout(); plt.show()
```

```python
# Box plot by biome
df.boxplot(column='TreeCoverLoss_ha', by='Biome', figsize=(9,5))
plt.suptitle(''); plt.title('Loss by Biome'); plt.ylabel('ha'); plt.xticks(rotation=20)
plt.tight_layout(); plt.show()
```

```python
# Scatter — loss vs carbon
for b, g in df.groupby('Biome'):
    a = g.groupby('Country').agg(L=('TreeCoverLoss_ha','sum'),C=('CarbonLoss_tCO2','sum')).reset_index()
    plt.scatter(a['L']/1e3, a['C']/1e6, label=b, alpha=0.75, s=55)
plt.xlabel('Loss (K ha)'); plt.ylabel('Carbon (Mt CO2)'); plt.title('Loss vs Carbon'); plt.legend()
plt.tight_layout(); plt.show()
```

## Key Findings
- Brazil, Indonesia & DR Congo lead total tree cover loss.
- Agricultural expansion is the #1 driver (~40%).
- Loss is accelerating — 2015–2023 significantly worse than 2001–2010.
- Loss and carbon emissions are near-perfectly correlated (r > 0.95).
